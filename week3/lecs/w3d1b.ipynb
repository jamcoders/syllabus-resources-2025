As we saw in the morning, some functions
take longer than others -- even our computer
isn't fast enough to do every function "instantly"

```python
def fib_recursive(n):
    if n <= 1:
        return n
    else:
        return fib_recursive(n-1) + fib_recursive(n-2)
```

```python
def fib_iterative(n):
    ans = [0]*(n+1)
    ans[1] = 1
    for i in range(2, n+1):
        ans[i] = ans[i-1] + ans[i-2]
    return ans[n]
```

```python
fib_iterative(10000)
```

```python
fib_iterative(40)
```

```python
start = time.time()
fib_iterative(36)
end = time.time()
print(end - start)
```

```python
start = time.time()
fib_recursive(36)
end = time.time()
print(end - start)
```
