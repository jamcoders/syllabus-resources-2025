## Boolean Operator Precedence: The Order of Operations!

Just like math has order of operations (PEMDAS), boolean expressions have their own rules!

### **Precedence Order (Highest to Lowest):**
1. **Parentheses** `()` - Always evaluated first
2. **Comparison Operators** `==`, `!=`, `<`, `>`, `<=`, `>=` - Evaluated next  
3. **NOT** `not` - Then logical NOT
4. **AND** `and` - Then logical AND
5. **OR** `or` - Finally logical OR

**Key Rule**: When operators have the same precedence, they evaluate **left to right**.

Let's see this in action with step-by-step examples!


````python
boolean_a = True
boolean_b = False
````

````python
not boolean_a or boolean_b
# not True or False
# False or False
# False
````

````python
True and True or False
````

ðŸ“Œ This means:
```python
not True and False  # is interpreted as (not True) and False
True or False and False  # is interpreted as True or (False and False)
```

ðŸ’¡ Always use parentheses `()` if you're unsure about the order â€” it makes your code clearer!

# Conditionals

## Motivation

Many times, we want the behavior of a program to depend on the value of a "thing". In other words, the program can change its behavior depending on a variable or user input.

As a simple example, let's say we want to assign "A" to the variable `letter_grade` if the `numeric_grade` is greater than 90. How can we do this?

````python
numeric_grade = 85

if numeric_grade > 90:
  print("Letter grade is A")
else:
  print("Letter grade is B")
  print("Congrats")
````

As we saw above, we need to manually assign the `letter_grade`. We currently don't know how to change the behavior of the program based on the `numeric_grade`!

We can do this using _conditionals_. I'll first show some examples and talk about the syntax.

Note: it's quite annoying to copy the code! We'll see how to fix this later

## Basic syntax for conditionals

The general syntax for conditionals is
```python
if <expression that evaluates to a boolean>:
    <expression>
else:
    <expression>
```

We will call `<expression that evaluates to a boolean>` the `<condition>`.

````python
1 == 2
````

````python
1 == 1
````

````python
if True:
  print("Hello")
else:
  print("Goodbye")
````

````python
if False:
  print("Hello")
else:
  print("Goodbye")
````

````python
num_1 = input("Enter a number:")
num_2 = input("Enter another number:")

if num_1 == num_2:
  print("The numbers are equal")
else:
  print("They are unequal")
````

````python
num_1 = int(input("Enter a number:"))
num_2 = int(input("Enter another number:"))

if num_1 > num_2:
  print("num1 is bigger")
else:
  print("num2 is bigger")
````

## More about conditionals


The expressions within the _body_ of the `if` or `else` can be complex, beyond a single statement.

Notice, the body is demarcated through *indentation*.

**Important note: the other branch is NOT executed**

Only one branch of the if or else is executed. If a branch is not taken and a variable is not defined, you will have an error

# String Functions

### Comparing Strings

Examples of how to compare strings

````python
name1 = input("name 1")
name2 = input("name 2")

print(name1, name2)
print(name1.lower(), name2.lower())
if name1.lower() == name2.lower():
  print("These are equal")
else:
  print("These are not equal")
````

````python
name1 = input("name 1").upper()
name2 = input("name 2").upper()

print(name1, name2)
if name1 == name2:
  print("These are equal")
else:
  print("These are not equal")
````
