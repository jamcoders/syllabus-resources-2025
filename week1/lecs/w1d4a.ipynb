## Return Values

```python
# Let's talk about return values.

# What does return do? It returns a value to the caller. That value
# is what the function call evaluates to.

# Write a function to compute the amount of tip to leave for a meal.

def calculate_tip(bill_amount, tip_percentage):
  tip_amount = bill_amount * tip_percentage
  return tip_amount
```

```python
patti_cost = 500
target_tip = 0.25
print("Patti cost:", patti_cost)
print("Target Tip:", target_tip)
```

```python
final_bill = calculate_tip(patti_cost, target_tip) + patti_cost
print(final_bill)

final_bill_smoothie = calculate_tip(700, 0.15) + 700
print("Smoothie final bill:", final_bill_smoothie)
```

Create a function that takes a list of strings and prints the lengths of all the strings in the list.

```python
my_strings = ["One", "good", "thing" "about music", " - ", "when it hits you," "you" "feel no pain."]

def my_string_len_func(lst):
    # Write your code here.
    for index in range(0, len(lst)):
      item = lst[index]
      print(len(item))

def my_string_len_func_second_attempt(lst):
  for item in lst:
    print(len(item))


def my_string_len_func_while(lst):
  index = 0
  while index < len(lst):
    item = lst[index]
    print(len(item))
    index = index + 1


my_string_len_func_while(my_strings)
```

__Question:__ Write a function `exponent` that takes as input two integers $x,y$ and returns $x^y$ without using `*` or `**`. It can use `multiply`

```python
def exponent(x,y):
    result = 1
    for i in range(y):
        result = multiply(result,x)
    return result
```

```python
print(exponent(2,3))
```

## Scope

The other big topic we'll talk about is scope. Scope refers to what variables are used or relevant in a particular block of code.

There are two kinds of scope:
 * Global Scope, which are variables that are available to use anywhere
 * Local Scope, which are variables that are available to use within a particular block

```python

num1 = 5
num2 = 12

def max_value(a, b):
  print("a: " + str(a) + ", b: " + str(b))
  if a > b:
    return a
  return b

print(max_value(num1, num2))
print("a: " + str(a) + ", b: " + str(b))
```

```python
def add_up(a, b, c):
  total = a + b + c
  return total

result = add_up(10, 5, 6)
print(result)
```
