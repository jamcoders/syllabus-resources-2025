(1) Let's write a function that takes a list of integers, and updates the list to flip the sign on each one.

````python
# Input: [5, 2, -6]
# Output: [-5, -2, 6]

# Input: []
# Output: []

# Destructive
def flips_destructive(input_list):
  for index in range(len(input_list)):
    current_item = input_list[index]
    input_list[index] = -current_item
  return input_list

# Non-destructive
def flips_nondestructive(input_list):
  new_list = []

  for item in input_list:
    new_list.append(-item)

  return new_list
````

````python
example_list = [5, 2, -6]
result = flips_nondestructive(example_list)

print(example_list)
print(result)
````

## Imports & Modules

````python
# First import we will use is the random library.
import random

# We can use random.randint(a, b) to get a random integer between a and b.
print(random.randint(1, 10))

# We can use random.choice(lst) to get a random element from a list.
print(random.choice([1, 2, 3, 4, 5]))

# We can use random.shuffle(lst) to shuffle a list.
lst = [1, 2, 3, 4, 5]
random.shuffle(lst)
print(lst)

# We can use random.random() to get a random float between 0 and 1.
print(random.random())

# We can use random.uniform(a, b) to get a random float between a and b.
print(random.uniform(1, 10))
````

````python
# Other cool imports

# We can use the math library to do math.
import math

# We can use math.sqrt(x) to get the square root of x.
print(math.sqrt(16))

# We can use math.pow(x, y) to get x to the power of y.
print(math.pow(2, 3))

````
