## While Loops

For loops were great for when you know how many times you want to iterate, or have a particular sequence.

Sometimes, you don't know how many times to iterate or over what sequence though...that is where while loops come in.

# While loop syntax

The syntax for while loops is:
```python
while <condition>:
    <expression>
```

As with `if` statements, the `<condition>` can be anything that evaluates to a boolean. The `<expression>` can be complex.

## Nested Loops

Loops can be nested inside one another, just like conditionals could be. Indentation becomes really important.

````python
# You can put loops inside loops
# ALWAYS make sure the variables for the loops are different

````

````python
n = 0
````

````python
# Example where nested loops are useful

# Example 1: Creating a Multiplication Table
print("=== MULTIPLICATION TABLE ===")
print("Why nested loops? For each number (outer), multiply by all numbers 1-5 (inner)")
print()

for i in range(1, 4):  # Outer loop: numbers 1, 2, 3
    print(f"Times table for {i}:")
    for j in range(1, 6):  # Inner loop: multiply by 1, 2, 3, 4, 5
        result = i * j
        print(f"  {i} × {j} = {result}")
    print()  # Blank line between tables
````
