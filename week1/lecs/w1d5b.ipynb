````python
size_of_pyramid = 4

for num in range(1, size_of_pyramid + 1):
  prefix_string = " " * (size_of_pyramid - num)
  core_pyramid = "# " * num
  print(prefix_string + core_pyramid)
````

````python
# Make a function that takes a pyramid size and prints that pyramid

def print_pyramid(size):
  print("I'm about to print a pyramid of size: " + str(size))
  for num in range(1, size + 1):
    prefix_string = " " * (size - num)
    core_pyramid = "# " * num
    print(prefix_string + core_pyramid)
````

````python
print_pyramid(-1)
````
