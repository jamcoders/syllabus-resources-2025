## Mutability

Lists are a *mutable* type, which means you can change the value itself without fully replacing it.



```python
names = ["Alex", "Manolis", "Frank", "Bruno"]
```

```python
print(names)
```

```python
names.append("Piyush")
```

```python
print(names)
```

```python
names.remove("Alex")
```

```python
print(names)
```

```python
names.sort()
```

```python
print(names)
```

## Functions that take lists as arguments

Mutable types are "passed by reference" when given to a function. So when the argument *binds* to the parameter name, they refer to listerally the same list.

As a result, any changes to the list made inside the function, are actually changing the list outside the function.

```python
to_do_list = []

def add_to_list(item, lst):
  lst.append(item) # This is a mutating change.

add_to_list("Shovel Snow", to_do_list)
print(to_do_list)
```

```python
add_to_list("wash dishes", to_do_list)
add_to_list("physics homework", to_do_list)
```

```python
def remove_from_list(item, lst):
  if not (item in lst):
    print("No can do!")
  else:
    lst.remove(item)

```

```python

print(to_do_list)
remove_from_list("Shovel Snow", to_do_list)
print(to_do_list)
```

```python
remove_from_list("attend jamcoders", to_do_list)
```
