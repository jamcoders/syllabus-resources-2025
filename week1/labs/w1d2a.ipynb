## Question 1: Style


**Code style** refers to the **unofficial rules for writing programs that make them easier for others to understand**. While the compiler doesn't care about style, it's incredibly important because **your future self and other people will need to read your code**.

We've covered some key conventions for good style:
*   **Variable names** must be in **"snakecase"** (all lowercase, words separated by underscores, like `num_people` or `is_raining`). They should be **descriptive** to clearly indicate what they represent, but not excessively long.

*   You should **use comments to describe what your code does** if it's not immediately obvious from variable names and operators. Comments are also useful for **documenting any assumptions** you've made.

*   **Indentation is crucial in Python for defining code blocks**. All lines indented immediately after a condition are part of that conditional's "body" and will **only be executed if the condition is true**. When lines revert to a previous indentation level, they are no longer part of that conditional's body and will execute every time. For nested conditionals, the inner block's indentation is one more level relative to its own condition statement.

**Example:**
```python
num_students = int(input("How many students? ")) # This line always executes


if num_students > 100: # This is the condition
    print("Wow, more than 100!") # This line is indented, part of the 'if' block [2]
    print("That's a lot!")       # This line is also indented, part of the 'if' block [2]

print("You said there are", num_students, "students.") # This line is not indented under the 'if', so it executes every time [3]

```

In this example, `"Wow, more than 100!"` and `"That's a lot!"` will only print if `num_students` is greater than 100. The last print statement, `"You said there are..."`, will always execute because its indentation level indicates it's outside the conditional block

You already know everything you need in order to be able to code your own choose your own adventure game, since you've worked with `input`, `print`, variables, and conditional statements.

Run the skeleton code below. Right now, this game only has 2 situations, each of which only has 2 options. Start off by updating the code to a new game theme. Get creative! Once you have it working, try to add another question.

If the students around you are also on Question 5, ask them to try out your game!

*Note:* The special character `\n` that you see adds a *new line* to your string and `\t` adds a *tab* to your string.

````python
# Input message
input_request = "Enter your choice: "                       # This text will be printed out when we get input from the user. Don't change it!

# Assume success at the start
success = True                                              # The 'success' variable keeps track of if you're still playing, or if you've failed.

# === Turn 1 === #
scene_1 = "Situation 1.\n"                                  # This gets printed out as the first situation. Change it to fit your game
question_1 = "Question 1?\n"                                # This gets printed out as the first question. Change it to fit your game
option_1a = "\tA: Option 1A.\n"                             # This gets printed out as option A for the first question. Change it to fit your game.
option_1b = "\tB: Option 1B.\n"                             # This gets printed out as option B for the first question. Change it to fit your game.

response_1a = "What happens if Option 1A is chosen.\n"      # This gets printed out if option A was chosen for question 1. Change it to fit your game.
response_1b = "What happens if Option 1B is chosen.\n"      # This gets printed out if option B was chosen for question 1. Change it to fit your game.

print(scene_1 + question_1 + option_1a + option_1b)         # This prints out the first situation, question, and options.
choice = input(input_request)                               # This asks the user to make a choice and saves their answer into the 'choice' variable

if choice == "A":
    print(response_1a)
    success = False                                         # Right now, choosing option A for question 1 makes you fail. You can change this to fit your game.
elif choice == "B":
    print(response_1b)

# === Turn 2 === #
# Only move on to the next turn if we have succeeded so far
if success:
    scene_2 = "Situation 2.\n"
    question_2 = "Question 2?\n"
    option_2a = "\tA: Option 2A.\n"
    option_2b = "\tB: Option 2B.\n"

    response_2a = "What happens if Option 2A is chosen.\n"
    response_2b = "What happens if Option 2B is chosen.\n"

    print(scene_2 + question_2 + option_2a + option_2b)
    choice = input(input_request)

    if choice == "A":
        print(response_2a)
    elif choice == "B":
        print(response_2b)
        success = False                                     # Right now, choosing option B for question 2 makes you fail. You can change this to fit your game.

# Ending
fail_text = "Game over :("                                  # This gets printed out if you lose. Change it to fit your game
success_text = "Game over :)"                               # This gets printed out if you win. Change it to fit your game

if success:
    print(success_text)
else:
    print(fail_text)
````
