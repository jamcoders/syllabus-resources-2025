# ```while``` Loop Recap
Recall from today's lecture that we were working on a guessing game, which used a ```for``` loop to keep track of the number attempts allowed. The code shown below allows the user to guess 5 times using the ```for i range(5)``` statement.

````python
secret_word = "bears"
for i in range(5):
    guess = input("Enter in a word guess:")
    if guess.lower() == secret_word:
        print("You got it!")
        break
    else:
        print("That's not right. Try again")
````

How would you change the number of attempts to 10? How about 20? Try it yourself!

<i>Hint: Take a look at the ```range``` function!</i>

````python
secret_word = "bears"
for i in range(5):
    guess = input("Enter in a word guess:")
    if guess.lower() == secret_word:
        print("You got it!")
        break
    else:
        print("That's not right. Try again.")
````

A ```for``` loop will work just fine if we know <b>ahead of time</b> the number of attempts. But, what if we don't know this number? Ideally, we want to keep repeating until the user guesses correctly.

Enter in the ```while``` loop! A ```while``` loop continues to repeat <i>indefinitely</i> as long as the condition expression evalues to ```True```.

In the code snippet below, as long as ```expression``` is evaluated to ```True```, all statements <b>indented</b> underneath the ```while``` loop repeats. The ```while``` loop will only end once ```expression``` is evaluated to ```False```.

```python
while expression:
    print("hello")
    print("world")
    print(":)")
```

Back to our game! We can simply replace the ```for``` loop with a ```while`` loop. This allows the user keep playing until they guess correctly!

````python
secret_word = "bears"
guess = input("Enter in a word guess: ")
while True:
    guess = input("Enter a word guess: ")
    if guess.lower() == secret_word:
        print("You got it!")
        break
    else:
        print("That's not right. Try again.")
````

Notice that our program exits because of the nested ```break``` statement. If we didn't ```break``` out, we would have an <b><i>infinite loop</i></b>, which would cause our program to repeat forever!

Try running the cell below, and notice that the program will keep repeating, even if we make the correct guess. End the cell by clicking on the <b>Stop</b> button.

````python
secret_word = "bears"
guess = input("Enter in a word guess: ")
while True:
    guess = input("Enter a word guess: ")
    if guess.lower() == secret_word:
        print("You got it!")
    else:
        print("That's not right. Try again.")
````

If we didn't want to use a ```break``` statement, we can alternatively change the conditional expression of our ```while``` loop by checking whether the user's input matches the secret word.

````python
secret_word = "bears"
guess = input("Enter a word guess: ")

while guess.lower() != "bears":
    print("That's not right. Try again.")
    guess = input("Enter a word guess: ")
print("You got it!")
````

# Nested Loops Recap
We can actually have nested ```for``` loops. This means we can place a ```for``` loop inside another ```for``` loop. These are used when we need to do something repeatedly inside another task that's also being repeated.

<b>Example 1:</b>
```python
for i in range(3):
    print("I'm the outer loop! Currently, i is", i)
    for j in range(4):
        print("Inner loop! j is", j)
```

```for i in range(3):``` is considered the outer loop. The loop nested inside is called the ```inner``` loop. In this example, the outer loop will repeat ```3``` times, and with each iteration, it will execute the inner loop ```4``` times.

First, run the code block below:

````python
for i in range(3):
    print("I'm the outer loop! Currently, i is", i)
    for j in range(4):
        print("Inner loop! j is", j)
````

Let's step through the code to see what exactly is happening.

First, ```i``` takes on the value ```0```, it will execute the inner loop ```4``` times. So, ```j``` takes on the values ```0```, ```1```, ```2```, ```3```, and prints ```"Inner Loop! j is, j"``` for each value.  

Next, ```i``` takes on the value ```1```, it will execute the inner loop ```4``` times. So, ```j``` takes on the values ```0```, ```1```, ```2```, ```3```, and prints ```"Inner Loop! j is, j"``` for each value.  

Lastly, ```i``` takes on the value ```2```, it will execute the inner loop ```4``` times. So, ```j``` takes on the values ```0```, ```1```, ```2```, ```3```, and prints ```"Inner Loop! j is, j"``` for each value.  


<b>Example 2</b>

We can also use nested ```for``` loops with lists!
```python
students = ["Mary", "Zara", "Edward", "Alex"]
school_materials = ["textbook", "lab equipment", "notebooks"]

for student in students:
    for material in school_materials:
        print(student, "received", material)
```

Here, ```for student in students:``` is the outer loop and the inner loop is ```for material in school_materials:```. We are assigning every ```student``` with the necessary school materials.

Run the code and examine the output:

## Bonus: Other Nested Loops

So far, we've seen nested ```for``` loops, but we can also use nested ```while``` loops. We can even nest a ```while``` loop inside a ```for``` loop, or nest a ```for``` loop inside a ```while``` loop!

Let's see how each of these would look for the two examples from the Nested Loop Recap Section:

<b>Example 1:</b>
```python
for i in range(3):
    print("I'm the outer loop! Currently, i is", i)
    for j in range(4):
        print("Inner loop! j is", j)
```

Nested ```while``` loops:

````python
# Nested while loop version of Example 1:
i = 0
while i < 3:
    print("I'm the outer loop! Currently, i is", i)
    j = 0
    while j < 4:
        print("Inner loop! j is", j)
        j += 1
    i += 1
````

Nested ```while``` loop inside a ```for``` loop:

````python
# Nested while loop inside a for loop of Example 1:
for i in range(3):
    print("I'm the outer loop! Currently, i is", i)
    j = 0
    while j < 4:
        print("Inner loop! j is", j)
        j += 1
````

Nested ```for``` loop inside a ```while``` loop:

````python
# Nested for loop inside a while loop of Example 1:
i = 0
while i < 3:
    print("I'm the outer loop! Currently, i is", i)
    for j in range(4):
        print("Inner loop! j is", j)
    i += 1
````

<b>Example 2</b>
```python
students = ["Mary", "Zara", "Edward", "Alex"]
school_materials = ["textbook", "lab equipment", "notebooks"]

for student in students:
    for material in school_materials:
        print(student, "received", material)
```

Nested ```while``` loops:

````python
students = ["Mary", "Zara", "Edward", "Alex"]
school_materials = ["textbook", "lab equipment", "notebooks"]

i = 0
while i < len(students):
    j = 0
    while j < len(school_materials):
        print(students[i], "received", school_materials[j])
        j += 1
    i += 1
````

Nested ```while``` loop inside a ```for``` loop:

````python
students = ["Mary", "Zara", "Edward", "Alex"]
school_materials = ["textbook", "lab equipment", "notebooks"]

for student in students:
    j = 0
    while j < len(school_materials):
        print(student, "received", school_materials[j])
        j += 1
````

Nested ```for``` loop inside a ```while``` loop:

````python
students = ["Mary", "Zara", "Edward", "Alex"]
school_materials = ["textbook", "lab equipment", "notebooks"]

i = 0
while i < len(students):
    for material in school_materials:
        print(students[i], "received", material)
    i += 1
````
