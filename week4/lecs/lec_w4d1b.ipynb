# Digital Skittles

What if we want to run our experiment on a thousand Skittles?

- That would be expensive ðŸ’¸

- But would make our dentist very happy ðŸ˜¬

We can run our experiment on the computer!

First, we need to to tell the computer about our cup:

```python
cup = ["red", "red", "red", "blue", "yellow", "green"]
```

Next, let's sample from it.
- How can we do something random on the computer?

- We can `import` it!

```python
from jamcoders.random import sample_from_list
```

```python
print(sample_from_list(cup)) # Sample from the cup
```

## Skittle-up ðŸ“ˆ

Let's make a cup with hundredsssss of Skittles...

- How many reds?
- How many blues?

```python
cup = ["red"] * 40
cup += ["blue"] * 100
cup += ["yellow"] * 69
```

### Running the experiment

Just like with our physical cups, we'll:
- Sample many times from our digital cup
- Try different sample sizes to see how our estimates improve

```python
from jamcoders.random import visualize
```

The `visualize` function will show us:
- What's in our cup ðŸ¥¤ (True distribution)
- What we saw when sampling (Sampled distribution)

The sampled distributions get closer to the true distribution as we take more samples

```python
sample_sizes = [50, 200, 1000, 5000]
visualize(cup, sample_sizes)
```
